```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

# Linear regression in multiple variables
  * we will try to predict house prices in an area on the basis of     area, number of bedrooms, and its age
  * price= M1* area + M2* bedrooms + M3*age + C(constant)

```python
df = pd.read_csv("homeprices.csv")
df
```

# One of the cells in bedrooms colums is empty.We should fill it up by taking a proper assumption

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms) #Empty cells updated with median value
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,15]])
```
